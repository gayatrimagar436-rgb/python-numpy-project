```python
num_cols=df.select_dtypes(include="number").columns
for col in num_cols:
   plt.figure(figsize=(6,4))
   sns.histplot(df[col],kde=True)
   plt.title(col)
   plt.title(col)
   plt.show()
```

```python
def plotting (var, num):
  plt.subplot(3,2,num)
  sns.histplot(df[var])

plotting("Release date",1)
plotting("Max resolution",2)
plotting("Low resolution",3)
plotting("Weight (inc. batteries)",4)
plotting("Dimensions",5)
plotting("Price",6)
plt.show()
```

```python

cat_cols = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(cat_cols)

# Plot count plots
for col in cat_cols:
    plt.figure(figsize=(10,5))

    sns.countplot(
        x=col,
        data=df,
        order=df[col].value_counts().index[:10],
        color="steelblue"# Top 10 categories
    )

    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.show()
```
